```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('INDIAvi.csv')
df.head()
```

```python
df.shape
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#handling null values
df['description']=df['description'].fillna('Not Available')
```

```python
df.head()
```

```python
df.describe()
```

```python
#showing first video of each category.
category_group=df.groupby('category_id')
category_group.first()
```

```python
df.corr()
```

```python
plt.bar(df.category_id,df.likes)
plt.title('Likes per Category')
plt.xlabel('category_id')
plt.ylabel('likes')
plt.show()
```

```python
sal = df.groupby(['category_id'], as_index=False)['views'].sum().sort_values(by='views', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sal, x = 'category_id',y= 'views')
```

```python
sal = df.groupby(['category_id'], as_index=False)['dislikes'].sum().sort_values(by='dislikes', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.scatterplot(data = sal, x = 'category_id',y= 'dislikes')
```

```python
sal = df.groupby(['channel_title'], as_index=False)['views'].sum().sort_values(by='views', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sal, x = 'channel_title',y= 'views')
```
